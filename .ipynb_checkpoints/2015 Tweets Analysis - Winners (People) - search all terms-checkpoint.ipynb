```python
def parseAward(award):
    award = re.split('\W+', award)
    award = [i for i in award if i not in stopwords]
    return award
```

```python
categories_dict = dict()
terms_counter = Counter()
for a in OFFICIAL_AWARDS_1315:
    terms = parseAward(a)
    for t in terms:
        terms_counter[t] += 1
    categories_dict[a] = terms
```

```python
def filter0(data, list1):
    result = []

    for tweet in data:
        if all(term in tweet.lower() for term in list1):
            result.append(tweet)

    return result
```

```python
def extractWinner(data, list1):
    #print(data[0])
    result = []
    
    tt = TweetTokenizer(strip_handles=True, reduce_len=True, preserve_case=True)
    translator = str.maketrans('', '', string.punctuation)

    # strip stopwords, punctuation 
    #punctuation = list(string.punctuation)
    remove_terms = ['#goldenglobes', 'golden globes', '#goldenglobe', 'golden globe', 'goldenglobes', 'goldenglobe', 'rt', 'golden', 'globe', 'globes']    
    stop = remove_terms + list1
    
    for tweet in data:
        #print(tweet)
        tweet = re.sub("\d+", "", tweet) #strip nums
        tweet = re.sub(r'http\S+', '', tweet) #strip urls
        tweet = re.sub(r'#\S+', '', tweet) #strip hashtags
        #tweet = re.sub(r'[^\w\s]', '', tweet) #strip non-alphanumeric characters
        tweet = tweet.translate(translator)
        tweet = tweet.split() #tokenize
        tweet = [term for term in tweet if term.lower() not in stop] #remove stop words
        result.append(tweet)
        
    #print(result[:20])
        
    bgrams = [];

    for tweet in result:
        if tweet:
            bigram = list(nltk.bigrams(tweet))
            #print(bigram[:10])
            for bg in bigram:
                if bool(re.search('([A-Z][a-z]+)', bg[0])) and bool(re.search('([A-Z][a-z]+)', bg[1])):
                    #print(bg[0], bg[1])
                    bgrams.append(bg)
            #tweet = re.findall('([A-Z][a-z]+)', tweet)
            #if tweet:
                #bgrams += list(nltk.bigrams(tweet))
  
    fdist = nltk.FreqDist(bgrams)
    #print(fdist)
    
    try:
        temp = fdist.most_common(1)[0][0]
        name = ' '.join(temp)
    except:
        name = "nothing here"
    
    return name
```

```python
def getActorWinners():
    
    c1 = filter0(data, category_dict['Best Actor – Motion Picture Drama'])
    c2 = filter0(data, category_dict['Best Actor – Motion Picture Musical or Comedy'])
    c3 = filter0(data, category_dict['Best Actress – Motion Picture Drama'])
    c4 = filter0(data, category_dict['Best Actress – Motion Picture Musical or Comedy'])
    c5 = filter0(data, category_dict['Best Supporting Actor – Motion Picture'])
    c6 = filter0(data, category_dict['Best Supporting Actress – Motion Picture'])
    c7 = filter0(data, category_dict['Best Actor – Television Series Drama'])
    c8 = filter0(data, category_dict['Best Actor – Television Series Musical or Comedy'])
    c9 = filter0(data, category_dict['Best Actor – Miniseries or Television Film'])
    c10 = filter0(data, category_dict['Best Actress – Television Series Drama'])
    c11 = filter0(data, category_dict['Best Actress – Television Series Musical or Comedy'])
    c12 = filter0(data, category_dict['Best Actress – Miniseries or Television Film'])
    c13 = filter0(data, category_dict['Best Supporting Actor – Series, Miniseries or Television Film'])
    c14 = filter0(data, category_dict['Best Supporting Actress – Series, Miniseries or Television Film'])
    director = filter0(data, category_dict['Best Director – Motion Picture'])
    screenplay = filter0(data, category_dict['Best Screenplay – Motion Picture'])
    original_score = filter0(data, category_dict['Best Original Score – Motion Picture'])
    cecil_demille = filter0(data, category_dict['Cecil B. DeMille Award for Lifetime Achievement in Motion Pictures'])
    #carol_burnett = filter0(data, ['carol', 'burnett'])
    
    winners = {
        'Best Actor – Motion Picture Drama' : extractWinner(c1, category_dict['Best Actor – Motion Picture Drama']),
        'Best Actor – Motion Picture Musical or Comedy' : extractWinner(c2, category_dict['Best Actor – Motion Picture Musical or Comedy']),
        'Best Actress – Motion Picture Drama' : extractWinner(c3, category_dict['Best Actress – Motion Picture Drama']),
        'Best Actress – Motion Picture Musical or Comedy' : extractWinner(c4, category_dict['Best Actress – Motion Picture Musical or Comedy']),
        'Best Supporting Actor – Motion Picture' : extractWinner(c5, category_dict['Best Supporting Actor – Motion Picture']),
        'Best Supporting Actress – Motion Picture' : extractWinner(c6, category_dict['Best Supporting Actress – Motion Picture']),
        'Best Actor – Television Series Drama' : extractWinner(c7, category_dict['Best Actor – Television Series Drama']),
        'Best Actor – Television Series Musical or Comedy' : extractWinner(c8, category_dict['Best Actor – Television Series Musical or Comedy']),
        'Best Actor – Miniseries or Television Film' : extractWinner(c9, category_dict['Best Actor – Miniseries or Television Film']),
        'Best Actress – Television Series Drama' : extractWinner(c10, category_dict['Best Actress – Television Series Drama']),
        'Best Actress – Television Series Musical or Comedy' : extractWinner(c11, category_dict['Best Actress – Television Series Musical or Comedy']),
        'Best Actress – Miniseries or Television Film' : extractWinner(c12, category_dict['Best Actress – Miniseries or Television Film']), 
        'Best Supporting Actor – Series, Miniseries or Television Film' : extractWinner(c13, category_dict['Best Supporting Actor – Series, Miniseries or Television Film']), 
        'Best Supporting Actress – Series, Miniseries or Television Film' : extractWinner(c14, category_dict['Best Supporting Actress – Series, Miniseries or Television Film']),
        'Best Director – Motion Picture' : extractWinner(director, category_dict['Best Director – Motion Picture']),
        'Best Screenplay – Motion Picture' : extractWinner(screenplay, category_dict['Best Screenplay – Motion Picture']),
        'Best Original Score – Motion Picture' : extractWinner(original_score, category_dict['Best Original Score – Motion Picture']),
        'Cecil B. DeMille Award for Lifetime Achievement in Motion Pictures' : extractWinner(cecil_demille, category_dict['Cecil B. DeMille Award for Lifetime Achievement in Motion Pictures'])
    }
    
    return winners
    
```
